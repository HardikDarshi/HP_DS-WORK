```python
#Basic data exploratory analysis
# 2-Find patterns
# Diff Complaint's Count
Complaints_Count=NYC['Complaint Type'].value_counts()
Complaints_Count
```

```python
#Visulising Complaint count
Complaints_Count.plot(kind='bar', figsize=(10,10))
plt.title('Count of complaints')
plt.xlabel('Type of complaints')
plt.ylabel('Count')
plt.show()
```

```python
#Status of complaints
Current_Status=NYC['Status'].value_counts(sort=True).plot(kind='bar',figsize=(5,8))
plt.show()
```

```python
#City wise Complaint Count
Citywisecomplaintcount=NYC['City'].value_counts()
Citywisecomplaintcount
```

```python
#Visualize City wise Complaint Count
Citywisecomplaintcount.plot(kind='barh', figsize=(17,17))
plt.xlabel('Count of complaints')
plt.ylabel('Name of City')
plt.title('City wise complaint count')
plt.show()
```

```python
#Basic data exploratory analysis
#3-Display the complaint type and city together
Group_NYC=NYC.groupby(['Complaint Type','City']).size().reset_index()
Group_NYC
```

```python
#Find major complaint types
#1-Find the top 10 complaint types 
Top10complaints=NYC['Complaint Type'].value_counts().nlargest(10)
Top10complaints
```

```python
#Find major complaint types
#2-Plot a bar graph of count vs. complaint types
Top10complaints.plot(kind='barh',figsize=(5,5),color=('green'))
plt.show()
```

```python
#Visualize the complaint types
#1-Display the major complaint types and their count

G_ComplaintType=NYC.groupby('Complaint Type')
Group_ComplaintType=G_ComplaintType.size().sort_values(ascending=False).nlargest(10).to_frame('Count').reset_index()
Group_ComplaintType
```

```python
plt.figure(figsize=(10,10))
plt.pie(Group_ComplaintType['Count'], labels=Group_ComplaintType['Complaint Type'],autopct="%1.1f%%")
plt.show()
```
